# Cleaning

```python
documents = []

for category in movie_reviews.categories():
    for file_id in movie_reviews.fileids(category):
        documents.append((movie_reviews.words(file_id),category))
        
documents[0:5]
```

```python
import random
random.shuffle(documents)

documents[0:5]
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()

from nltk.corpus import wordnet
```

```python
def get_simple_pos(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    return wordnet.NOUN
```

```python
word = "greater"
from nltk import pos_tag

print(pos_tag(word)) # not passed as array

print(pos_tag([word])) # should be passed as array
```

```python
from nltk.corpus import stopwords
stops = stopwords.words('english')
print(len(stops))

import string
punc = list(string.punctuation)
stops += punc
print(len(stops))
```

```python
# a function for converting words to their root words

def clean_review(words):
    output_words = []
    for word in words:
        if word.lower() not in stops:
            pos_t = pos_tag([word])
            new_word = lemmatizer.lemmatize(word,pos = get_simple_pos(pos_t[0][1])) # refer to word greater dry run if doubts over [0][1]
            output_words.append(new_word)
            
    return output_words
```

```python
documents[0]
```

```python
documents = [(clean_review(words), category) for words,category in documents]
```

```python
documents[0]
```

# Building Features

```python
training_data = documents[0:1500]
testing_data = documents[1500:]
```

```python
all_words = []
for doc in training_data:
    all_words += doc[0]
```

```python
import nltk
```

```python
words_frequency = nltk.FreqDist(all_words)
words_frequency
```

```python
top_words = words_frequency.most_common(3000)
print(top_words)

features = [i[0] for i in top_words]
features
```

```python
# create a dictionary for each document whether the words come in document or not
# because to use nltk naive bayes, format should be a tuple of (dictionary,category)

def get_feature_dic(words):
    current_features = {}
    words_set = set(words)
    for w in features:
            current_features[w] = w in words_set
    return current_features
```

```python
# modified data acc to nltk
training_data_modified = [(get_feature_dic(doc),category) for doc,category in training_data]
testing_data_modified = [(get_feature_dic(doc),category) for doc,category in testing_data]
```

# Applying NaiveBayesClassifier

```python
from nltk import NaiveBayesClassifier
```

```python
clf = NaiveBayesClassifier.train(training_data_modified)
```

```python
nltk.classify.accuracy(clf,testing_data_modified)
```

```python
nltk.classify.accuracy(clf,training_data_modified)
```
